### 01) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt
%matplotlib inline
values=[305,201,405,350,200]
l=['renii','krisha','amee','yashvi','devanshi']
c=['r','b','g','c','m']
e=[0.2,0,0,0,0]
plt.subplot(1,2,1)
plt.pie(values,colors=c,shadow=True,labels=l,autopct='%1.1f%%')
plt.subplot(1,2,2)
plt.pie(values,colors=c,shadow=True,labels=l,autopct='%1.1f%%',explode=e)
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ['r','g','y','m']
plt.bar(x,y,color=c,edgecolor="b")
def addLabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.title("popularity of languages")
plt.xlabel("languages")
plt.ylabel("ratings")
addLabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
c = ["r","b","g","r","b"]
width = [1,2,3,4,5]
plt.scatter(x,y,color=c,marker="*",linewidths=width)
plt.xlabel("x")
plt.ylabel("y")
plt.title("star")
plt.show()
```
